```python
import pandas as pd
import numpy as np

# Fetch the Boston housing dataset from the original source
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# Extract features into a DataFrame
# The feature names are not directly available from the raw data.
# I will use a placeholder or infer them if possible for now.
# For simplicity, I'll use generic column names as they are not explicitly provided in the raw data loading method.
# The original dataset had 13 features. We can label them X0 to X12.
X_boston = pd.DataFrame(data, columns=[f'X{i}' for i in range(data.shape[1])])

# Extract target variable into a Series
y_boston = pd.Series(target, name='MEDV')

print("Boston Housing Dataset loaded successfully.")
print("Features (X_boston) shape:", X_boston.shape)
print("Target (y_boston) shape:", y_boston.shape)
print("\nFirst 5 rows of X_boston:")
print(X_boston.head())
print("\nFirst 5 values of y_boston:")
print(y_boston.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_boston, y_boston, test_size=0.2, random_state=42)

# Instantiate the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

print("Linear Regression model trained successfully.")
print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print the evaluation metrics
print(f"R-squared (R²): {r2:.3f}")
print(f"Mean Squared Error (MSE): {mse:.3f}")
print(f"Mean Absolute Error (MAE): {mae:.3f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.3f}")
```

```python
import pandas as pd

# Access the coefficients of the trained linear regression model
coefficients = model.coef_

# Access the intercept of the trained linear regression model
intercept = model.intercept_

# Create a pandas Series to store the coefficients with feature names as index
feature_coefficients = pd.Series(coefficients, index=X_boston.columns)

print("Model Intercept:", intercept)
print("\nFeature Coefficients:")
print(feature_coefficients)
```

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame for features
X_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# Create a Series for the target variable
y_iris = pd.Series(iris.target, name='species')

print("Iris Dataset loaded successfully.")
print("Features (X_iris) shape:", X_iris.shape)
print("Target (y_iris) shape:", y_iris.shape)
print("\nFirst 5 rows of X_iris:")
print(X_iris.head())
print("\nFirst 5 values of y_iris:")
print(y_iris.head())
```

```python
from sklearn.model_selection import train_test_split

# Split the Iris dataset into training and testing sets
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)

print("Iris dataset split into training and testing sets.")
print(f"Training features shape: {X_train_iris.shape}")
print(f"Testing features shape: {X_test_iris.shape}")
print(f"Training target shape: {y_train_iris.shape}")
print(f"Testing target shape: {y_test_iris.shape}")
```

```python
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Using 'liblinear' solver for multiclass classification as it handles L1 and L2 penalties
# 'multi_class' is set to 'ovr' (one-vs-rest) as the dataset has more than two classes
# 'max_iter' is increased to ensure convergence, if needed
model_iris = LogisticRegression(solver='liblinear', multi_class='ovr', max_iter=200, random_state=42)

# Fit the model to the training data
model_iris.fit(X_train_iris, y_train_iris)

print("Logistic Regression model for Iris trained successfully.")
```

**Reasoning**:
The previous code block generated a FutureWarning indicating that the 'multi_class' parameter with 'ovr' strategy is deprecated in `LogisticRegression` since version 1.5. To address this, I will explicitly use `OneVsRestClassifier` to handle multiclass classification, as recommended by the warning.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

# Instantiate the Logistic Regression model
# Using 'liblinear' solver for multiclass classification as it handles L1 and L2 penalties
# 'max_iter' is increased to ensure convergence, if needed
logistic_regression_base = LogisticRegression(solver='liblinear', max_iter=200, random_state=42)

# Wrap the LogisticRegression model with OneVsRestClassifier for explicit 'ovr' strategy
model_iris = OneVsRestClassifier(logistic_regression_base)

# Fit the model to the training data
model_iris.fit(X_train_iris, y_train_iris)

print("Logistic Regression model for Iris trained successfully using OneVsRestClassifier.")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Make predictions on the test set
y_pred_iris = model_iris.predict(X_test_iris)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test_iris, y_pred_iris)
classification_rep = classification_report(y_test_iris, y_pred_iris)

# Print the evaluation metrics
print(f"Accuracy: {accuracy:.3f}")
print("\nClassification Report:\n", classification_rep)
```

### Subtask:

**Reasoning**:

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

def plot_learning_curve(estimator, title, X, y, axes=None, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5),
                        scoring='neg_mean_squared_error'):
    """
    Generate 3 plots: the test and training learning curve, the training
    samples vs. fit times curve (if fit_times_plot is True),
    and the fit times vs. score curve (if fit_times_plot is True).
    """
    if axes is None:
        _, axes = plt.subplots(1, 1, figsize=(10, 5))

    axes.set_title(title)
    if ylim is not None:
        axes.set_ylim(*ylim)
    axes.set_xlabel("Training examples")
    axes.set_ylabel("Score (MSE)")

    train_sizes, train_scores, test_scores, fit_times, _ = \
        learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                       train_sizes=train_sizes, scoring=scoring,
                       return_times=True)

    train_scores_mean = -np.mean(train_scores, axis=1)  # Convert to positive MSE
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = -np.mean(test_scores, axis=1)    # Convert to positive MSE
    test_scores_std = np.std(test_scores, axis=1)

    # Plot learning curve
    axes.grid()
    axes.fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
    axes.fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1,
                         color="g")
    axes.plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
    axes.plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Cross-validation score")
    axes.legend(loc="best")

    return plt

# Prepare figure for multiple plots
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

# 1. Learning Curve for Linear Regression
plot_learning_curve(model, "Learning Curve for Linear Regression", X_train, y_train, axes=axes[0], cv=5)

# 2. Learning Curve for Lasso Regression
plot_learning_curve(lasso_model, "Learning Curve for Lasso Regression", X_train, y_train, axes=axes[1], cv=5)

# 3. Learning Curve for Ridge Regression
plot_learning_curve(ridge_model, "Learning Curve for Ridge Regression", X_train, y_train, axes=axes[2], cv=5)

plt.tight_layout()
plt.show()

print("Learning curves for Linear, Lasso, and Ridge Regression models generated and plotted.")
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

def plot_learning_curve_classification(estimator, title, X, y, axes=None, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5),
                        scoring='accuracy'):
    """
    Generate 3 plots: the test and training learning curve, the training
    samples vs. fit times curve (if fit_times_plot is True),
    and the fit times vs. score curve (if fit_times_plot is True).
    Adapted for classification scoring.
    """
    if axes is None:
        _, axes = plt.subplots(1, 1, figsize=(10, 5))

    axes.set_title(title)
    if ylim is not None:
        axes.set_ylim(*ylim)
    axes.set_xlabel("Training examples")
    axes.set_ylabel(f"Score ({scoring.capitalize()})")

    train_sizes, train_scores, test_scores, fit_times, _ = \
        learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                       train_sizes=train_sizes, scoring=scoring,
                       return_times=True)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    # Plot learning curve
    axes.grid()
    axes.fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
    axes.fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1,
                         color="g")
    axes.plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
    axes.plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Cross-validation score")
    axes.legend(loc="best")

    return plt

# Prepare figure for multiple plots
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

# 1. Learning Curve for Original Logistic Regression
plot_learning_curve_classification(model_iris, "Learning Curve: Logistic Regression (Iris)",
                                   X_train_iris, y_train_iris, axes=axes[0], cv=5, scoring='accuracy')

# 2. Learning Curve for L1 Regularized Logistic Regression
plot_learning_curve_classification(model_iris_l1, "Learning Curve: L1 Logistic Regression (Iris)",
                                   X_train_iris, y_train_iris, axes=axes[1], cv=5, scoring='accuracy')

# 3. Learning Curve for L2 Regularized Logistic Regression
plot_learning_curve_classification(model_iris_l2, "Learning Curve: L2 Logistic Regression (Iris)",
                                   X_train_iris, y_train_iris, axes=axes[2], cv=5, scoring='accuracy')

plt.tight_layout()
plt.show()

print("Learning curves for Logistic Regression models on Iris dataset generated and plotted.")
```
