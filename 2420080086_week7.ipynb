```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons , make_circles
```

```python
x_moon , y_moon = make_moons(n_samples=200 , noise=0.2)
plt.scatter(x_moon[:,0] , x_moon[:,1] , c=y_moon)
plt.title("Moon Dataset")
plt.show()
```

```python
x_circle , y_circle = make_circles(n_samples=200 , noise=0.1  , factor = 0.5)
plt.scatter(x_circle[:,0] , x_circle[:,1] , c=y_circle)
plt.title("Circles Datset")
plt.show()
```

```python
from sklearn.svm import SVC

svm_learn = SVC(kernel='linear' , C=1)
svm_learn.fit(x_circle , y_circle)
```

```python
svm_poly2 = SVC(kernel = 'poly' , degree = 2)
svm_poly2.fit(x_circle , y_circle)
svm_poly4= SVC(kernel = 'poly' , degree = 3)
svm_poly4.fit(x_circle , y_circle)
svm_poly6 = SVC(kernel = 'poly' , degree = 4)
svm_poly6.fit(x_circle , y_circle)


```

```python
svm_rbf1 = SVC(kernel = 'rbf' , gamma = 0.5)
svm_rbf1.fit(x_circle , y_circle)
svm_rbf2 = SVC(kernel = 'rbf' , gamma = 1)
svm_rbf2.fit(x_circle , y_circle)
svm_rbf3 = SVC(kernel = 'rbf' , gamma = 5)
svm_rbf3.fit(x_circle, y_circle)
```

```python
svm_sigmoid =SVC(kernel='sigmoid')
svm_sigmoid.fit(x_circle,y_circle)
```

```python
def plot_svm(model , X , y , title):
  xx , yy = np.meshgrid(
      np.linspace(X[:,0].min() - 1 , X[:,0].max() + 1 , 100),
      np.linspace(X[:,1].min() - 1 , X[:,1].max() + 1 , 100)
  )
  Z = model.predict(np.c_[xx.ravel() , yy.ravel()])
  Z = Z.reshape(xx.shape)
  plt.contourf(xx , yy , Z , alpha = 0.4)
  plt.scatter(X[:,0] , X[:,1] , c=y)
  plt.scatter(model.support_vectors_[:,0],
              model.support_vectors_[:,1],
              s=100 , edgecolor='k' , facecolors='none')
  plt.title(title)
  plt.show()
```

```python
plot_svm(svm_learn , x_circle , y_circle , "Linear Kernel")
plot_svm(svm_poly2 , x_circle , y_circle , "Polynomial Kernel (d=2)")
plot_svm(svm_rbf2 , x_circle , y_circle , "RBF Kernel (gamma=1)")
plot_svm(svm_sigmoid , x_circle , y_circle , "Sigmoid Kernel")
```

```python
print("Linear Accuracy:", svm_learn.score(x_circle, y_circle))
print("Poly d=2 Accuracy:", svm_poly2.score(x_circle, y_circle))
print("RBF gamma = 1 Accuracy:", svm_rbf1.score(x_circle, y_circle))
print("Sigmoid Accuracy:", svm_sigmoid.score(x_circle, y_circle))
```
